```python
import matplotlib.pyplot as plt
# pie.plt(values,labels="label_list",color="color_list",autopct="%1.1f%%")

regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['lightgreen','skyblue','coral','gold'])
plt.title("Revenue Contrubition by Regions")
plt.show()
```
