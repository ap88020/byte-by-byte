```python
#plt.bar(x,hights,color='color name',width=value , label='label')

import matplotlib.pyplot as plt
product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product,sales,color='orange',label="sales of 2025")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Comparision")
plt.legend()
plt.show()
```
