```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1) # 1 -> row , column -> column , 1 -> positon
plt.plot([1,2,3,4],[10,15,20,25],color="blue")
plt.title("Line Chart")
plt.grid()

plt.subplot(1,2,2) # 1->row , column->2 , 2-> positon
plt.bar([1,2,3,4],[10,15,20,25],color="green")
plt.title("Bar Chart")
plt.grid()

# plt.tight_layout()
plt.show()
```
