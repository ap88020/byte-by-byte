```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4],[10,15,20,12] , color="blue", marker="^" , label="section A");
plt.scatter([1,2,3,4],[20,25,30,22] , color="yellow", marker="*" , label="section A");
plt.title("Students Data")
plt.xlabel("Hours of studies")
plt.ylabel("scores of exames")
plt.legend()
plt.grid(True)
plt.savefif("scatter_graph.png",dpi=300,bbox_inches="tight")
plt.show()
```
